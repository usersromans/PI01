```python
### Importamos las librerias.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Carga de datos
df_steam_games = pd.read_csv('./output_steam_games.csv')

```

```python
df_steam_games.info()
df_steam_games.head(40)
```

```python
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'], errors='coerce')

# Extraer solo el año y crear una nueva columna 'year'
df_steam_games['year'] = df_steam_games['release_date'].dt.year
```

```python
df_steam_games.info()
df_steam_games.head(40)
```

```python
# Lista de columnas de géneros
genre_columns = ['Action', 'Adventure', 'Animation_&amp;_Modeling', 'Audio_Production', 'Casual', 'Design_&amp;_Illustration', 'Early_Access', 'Education', 'Free_to_Play', 'Indie', 'Massively_Multiplayer', 'Photo_Editing', 'RPG', 'Racing', 'Simulation', 'Software_Training', 'Sports', 'Strategy', 'Utilities', 'Video_Production', 'Web_Publishing', 'Accounting', 'Design', 'Photo', 'Web', 'Software', 'Audio', 'Early', 'Animation', 'Video', 'Free', 'Massively']

# Conteo de juegos por género
genre_counts = df_steam_games[genre_columns].sum().sort_values(ascending=False)

# Gráfico de barras para géneros
genre_counts.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.title('Number of Games per Genre')
plt.show()
```

```python
# Histograma de precios
sns.histplot(df_steam_games['price'].dropna(), bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Game Prices')
plt.show()
```

```python
# Convertir 'release_date' a datetime
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'], errors='coerce')

# Conteo de juegos por año
games_per_year = df_steam_games['release_date'].dt.year.value_counts().sort_index()

# Gráfico de líneas para juegos por año
games_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.title('Games Released Over Years')
plt.show()
```

```python
df_steam_games.to_csv('output_steam_games.csv', index=False)
```

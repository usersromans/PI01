```python
import pandas as pd
import json

file = './output_steam_games.json'

data = []
with open(file, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            # Usa json.loads para interpretar la línea como un diccionario de Python
            data_dict = json.loads(line)
            data.append(data_dict)
        except json.JSONDecodeError as e:
            print(f"Error al interpretar la línea: {e}")
            # Puedes elegir continuar en caso de error o interrumpir el bucle
            continue

# Convierte la lista de diccionarios a un DataFrame
df = pd.DataFrame(data)

```

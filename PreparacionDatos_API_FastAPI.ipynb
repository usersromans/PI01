```python
import pandas as pd 

path = './output_steam_games.csv'

df_info_plays = pd.read_csv(path)

df_info_plays.info()
df_info_plays.head()
```

```python
### extraemos las colunas que vamos a utilizar para las funciones en un nuevo df 

data_necesaria = ['id', 'publisher','title', 'year', 'price', 'Action','Adventure','Animation_&amp;_Modeling','Audio_Production','Casual','Design_&amp;_Illustration','Early_Access','Education','Free_to_Play','Indie','Massively_Multiplayer','Photo_Editing','RPG','Racing','Simulation','Software_Training','Sports','Strategy','Utilities','Video_Production','Web_Publishing','Accounting','Design','Photo','Web','Software','Audio','Early','Animation','Video','Free','Massively'  ]

df_plays = df_info_plays[data_necesaria].copy()

df_plays.info()
df_plays.head()
```

```python
import pandas as pd

# Supongamos que tienes un DataFrame df_plays con las columnas mencionadas
# Define una función para ajustar una cadena según los requisitos
def ajustar_cadena(cadena):
    # Convierte la cadena a string por si acaso
    cadena = str(cadena)
    
    # Elimina espacios, puntos y comas solo al principio y al final
    cadena = cadena.strip(" ,.")

    # Intenta convertir a float y luego a int, si es posible
    try:
        # Primero intenta convertir a float
        flotante = float(cadena)
        # Si es exitoso, convierte el float a int
        cadena = str(int(flotante))
    except ValueError:
        # Si hay un error, mantiene la cadena como está
        pass

    # Devuelve la cadena ajustada
    return cadena

# Aplica la función a las columnas que deseas ajustar
columnas_ajustar = ['id', 'publisher', 'title', 'year']

for columna in columnas_ajustar:
    df_plays[columna] = df_plays[columna].apply(ajustar_cadena)

# Ahora, el DataFrame df_plays tiene valores ajustados en las columnas especificadas
print(df_plays)

```
